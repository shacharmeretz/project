## split and encoder

```python
def split(df,label_col):
  labels_df=df[label_col]
  data_df= df.drop(label_col, axis='columns')
  return data_df,labels_df
```

```python
def encoder(df):
  #remove cols that are not number
  cols = df.select_dtypes(exclude=['number']).columns
  for col in cols:
    le = preprocessing.LabelEncoder()
    le.fit(df[col])
    df[col] = le.transform(df[col])
  return df
```

##model

```python
def create_model(opt,active_name):
    model = tf.keras.models.Sequential([
        tf.keras.layers.Flatten(),                         
        tf.keras.layers.Dense(64, activation=active_name),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dense(64, activation=active_name),
        tf.keras.layers.BatchNormalization(),
        tf.keras.layers.Dense(2, activation='softmax')
    ])

    model.compile(optimizer=opt,
                    loss='sparse_categorical_crossentropy',
                    metrics=['accuracy'])

    return model
```

```python
# evaluate a specific number of members in an ensemble
def evaluate_model(model,testX, testy):

  start_predict = timeit.default_timer()

  yhat = model.predict(testX) 
  yhat = yhat.argmax(axis=-1)
  end_predict = timeit.default_timer()
  total_time_predict=end_predict-start_predict
  print('total_time_predict: %s' % total_time_predict)

  len_of_predict=len(testX)
  time_infernce_for_1000_instance=(total_time_predict/len_of_predict)*1000
  print('time_infernce_for_1000_instance: %s' % time_infernce_for_1000_instance)

  
  acc=accuracy_score(testy, yhat)
  print('acc: %s' % acc)

  model_precision, model_recall, _ = precision_recall_curve(testy, yhat)
  model_auc_pr = auc( model_recall,  model_precision)
  model_precision=model_precision[0]
  print('model_precision: %s' % model_precision)
  print('model_auc_pr: %s' % model_auc_pr)
  
  y_hat_prob=model.predict(testX)
  
  model_auc_roc = roc_auc_score(testy,  model.predict_proba(testX)[:, 1])
  print('model_auc: %s' % model_auc_roc)

  returned  = metrics.confusion_matrix(testy, yhat).ravel()
  if len(returned) == 4:
    TN, FP, FN, TP  = returned
    confusion_matrix_tpr= round(TP/(TP+FN), 3)
    confusion_matrix_fpr= round(FP/(FP+TN), 3)
    print('confusion_matrix_tpr: %s' % confusion_matrix_tpr)
    print('confusion_matrix_fpr: %s' % confusion_matrix_fpr)
  else:
    confusion_matrix_tpr=None
    confusion_matrix_fpr=None
  return acc,model_precision,model_auc_pr,model_auc_roc,confusion_matrix_fpr,confusion_matrix_tpr,time_infernce_for_1000_instance
```

```python
def sgd_model(data_df,label_class,dataset_name,df_total,is_swa):
  global X_train, y_train, X_test, y_test
  kfold_model = KFold(n_splits=10)
  stratified_kfold = StratifiedKFold(n_splits=10 , shuffle = True , random_state = 4)
  round_kfold = 1
  activation_model=None
  learning_rate=None
  for train_index, val_index in kfold_model.split(np.zeros(len(data_df)),data_df):
    print("Start FOLD Number {}".format(round_kfold))
    max=0
    ok = OptKeras(study_name='GridSearch',
              monitor='val_acc',
              direction='maximize',
              random_grid_search_mode=True)
    stratified_kfold = StratifiedKFold(n_splits=3 , shuffle = True , random_state = 4)
    kfold_model = KFold(n_splits=3)
    for params_train_index, params_val_index in kfold_model.split(np.zeros(len(data_df)),data_df):
        training = data_df.iloc[params_train_index]
        test = data_df.iloc[params_val_index]
        X_train, y_train=split(data_df,label_class)
        X_test, y_test=split(test,label_class)
        ok.random_grid_search(objective,n_trials=int(50))
        value_fold=ok.best_trial.value
        if value_fold>max:
          max=value_fold
          activation_model=ok.best_trial.params['activation']
          learning_rate=ok.best_trial.params['learning_rate']
    if activation_model is None or learning_rate is None:
      pass
    print("activation_model",activation_model)
    print("learning_rate",learning_rate)
    training = data_df.iloc[train_index]
    test = data_df.iloc[val_index]
    X_train, y_train=split(data_df,label_class)
    X_test, y_test=split(test,label_class)

    checkpoint = tf.keras.callbacks.ModelCheckpoint(os.path.join(checkpoint_path,'sgd_model.h5'), monitor='val_accuracy', verbose=1, save_best_only=True, mode='max')
    sgd = tf.keras.optimizers.SGD(learning_rate)
    model = create_model(sgd,activation_model)

    n_epochs = 50
    start_train = timeit.default_timer()

    if is_swa:
      swa = SWA(start_epoch=10, lr_schedule='cyclic', swa_lr=0.0001,   swa_lr2=0.003,   swa_freq=3,  batch_size=32, verbose=1)
      model.fit(X_train, y_train, epochs=n_epochs,validation_data=(X_test, y_test), callbacks=[checkpoint,swa])
    else:
      model.fit(X_train, y_train, epochs=n_epochs,validation_data=(X_test, y_test), callbacks=[checkpoint])

    end_train = timeit.default_timer()
    total_time_train=end_train-start_train
    print('total_time_train: %s' % total_time_train)

    acc,model_precision,model_auc_pr,model_auc_roc,confusion_matrix_fpr,confusion_matrix_tpr,time_infernce_for_1000_instance=evaluate_model(model,X_test,y_test)

    if is_swa:
      new_row={"Dataset Name":dataset_name,
                   "Algorithm Name":"swa_classifier",
                   "Cross Validation":round_kfold,
                   "Hyper-Parameters activation":activation_model,
                   "Hyper-Parameters learning_rate":learning_rate,
                   "Accuracy":acc,
                   "TPR":confusion_matrix_tpr,
                   "FPR":confusion_matrix_fpr,
                   "Precision":model_precision,
                   "AUC":model_auc_roc,
                   "PR-Curve":model_auc_pr,
                   "Training Time":total_time_train,
                   "Inference Time":time_infernce_for_1000_instance,
                   }
    else:
        new_row={"Dataset Name":dataset_name,
                   "Algorithm Name":"sgd_classifier",
                   "Cross Validation":round_kfold,
                   "Hyper-Parameters activation":activation_model,
                   "Hyper-Parameters learning_rate":learning_rate,
                   "Accuracy":acc,
                   "TPR":confusion_matrix_tpr,
                   "FPR":confusion_matrix_fpr,
                   "Precision":model_precision,
                   "AUC":model_auc_roc,
                   "PR-Curve":model_auc_pr,
                   "Training Time":total_time_train,
                   "Inference Time":time_infernce_for_1000_instance,
                   }
    df_total=df_total.append(new_row, ignore_index=True)
    with open('/content/drive/MyDrive/ML/df.csv', 'a') as f_object:
      writer_object = writer(f_object)
      if is_swa:
        writer_object.writerow([0,dataset_name,"swa_classifier",round_kfold,activation_model,learning_rate,acc,confusion_matrix_tpr,confusion_matrix_fpr,model_precision,model_auc_roc,total_time_train,time_infernce_for_1000_instance])
      else:
        writer_object.writerow([0,dataset_name,"sgd_classifier",round_kfold,activation_model,learning_rate,acc,confusion_matrix_tpr,confusion_matrix_fpr,model_precision,model_auc_roc,total_time_train,time_infernce_for_1000_instance])
      f_object.close()
    round_kfold += 1
  return df_total
```
